```
x = range(100)
```

```
plt.plot(np.square(x))      #  Построим график x в квадрате.
plt.show()
```

```
x = np.linspace(0, 20, 100)     #  Создаёт массив из 100 значений равномерно распределенных от 0 до 20.
```

```
plt.figure(figsize=(15,5))      #  Меняет размер изображения.
plt.plot(np.sin(x), '^r:')      #  Маркер '^r:' меняет голубую линию на красные треугольники с точками.
plt.title('Заголовок')      #  Добавляет заголовок.
plt.ylabel('sin(x)')      #  Подписывает ось y.
plt.xlabel('x')      #  Подписывает ось x.
plt.show()
```

```
plt.figure(figsize=(15,5))
plt.plot(np.sin(x), 'r--')      #  Маркер 'r--' меняет голубую линию на прерывистую красную линию.
plt.title('Заголовок')
plt.ylabel('sin(x)')
plt.xlabel('x')
plt.grid(True)      #  Добавляет вспомогательную сетку на рисунок.
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.hist(np.sin(x))      #  Строит гистрограмму частот.
plt.show()
```

```
plt.figure(figsize=(10,6))
x = np.linspace(0, 20, 100)
y = np.random.randint(0, 100, 100)
z = np.abs(np.random.randn(100)) * 100
i = x + 20 * np.random.randn(100)

plt.scatter(x, i, c=y, s=z)
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.show()
```

# Цветовые карты, работа с изображениями как с массивами данных.

## Цветовые карты — это один из видов представление информации, основанный на сопоставлении одного цвета одному или нескольким значениям.


```
x = np.random.rand(20,20)     # Создадим массив из случайных чисел.

# Построим цветовую карту массива.
# Чем ближе значение элемента массива к единице, тем светлее его отображение на цветовой карте.
plt.figure(figsize=(10,6))
plt.pcolormesh(x,cmap='hot')      
plt.show()
```

## А теперь давайте взглянем на изображение в виде массива.

```
from sklearn.datasets import load_digits
```

```
#  Из библиотеки sklearn.datasets скачаем датасет с рукописными цифрами от 0 до 9.
digits = load_digits()
```

```
#  Датасет содержит 1797 изображений рукописных цифр от 0 до 9.
#  Размер изображения 8*8 пикселей
print(digits.data.shape)      
```

```
# Посмотрим на цифровое представление изображения цифры 0.
digits.images[0]
```

```
# Теперь посмотрим на изображение.
plt.figure(figsize=(6,6))
plt.matshow(digits.images[0], fignum=1, aspect='auto') 
plt.show() 
```

```
# Посмотрим на цифровое представление изображения цифры 1.
digits.images[200]
```

```
# Посмотрим на цифровое представление изображения цифры 1.
plt.figure(figsize=(6,6))
plt.matshow(digits.images[200], fignum=1, aspect='auto') 
plt.show() 
```
